### Projection of the mapping data to 2D plots

```python
import pandas as pd
import numpy as np
%matplotlib notebook
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from matplotlib.ticker import FormatStrFormatter
import numpy as np

```

```python
df1 = pd.read_csv('Mapping_0809_RUN1.csv')
df2 = pd.read_csv('Mapping_0809_RUN2.csv')
df3 = pd.read_csv('Mapping_0809_RUN3.csv')
df4 = pd.read_csv('Mapping_0809_RUN4.csv')

df_all0 = df1.append(df2)
df_all1  = df_all0.append(df3)
df_all  = df_all1.append(df4)
```

```python
u_4 =  df4.u.unique()
w_4 = df4.w.unique()
v_4 = df4.v.unique()
print u_4
print v_4
print w_4
```

```python
fig0, ax0 = plt.subplots(2,3, figsize=(11,6))

for i in range(2):
    for j in range(len(w_4)):
            ax0[0,0].plot(df4[(df4.u==u_4[i])& (df4.w==w_4[j])].v, df4[(df4.u==u_4[i])& (df4.w==w_4[j])].B_u, 
                          '.-', lw=0.8, label='u=%.2f,w=%.2f'%(u_4[i], w_4[j]))
            ax0[0,1].plot(df4[(df4.u==u_4[i])& (df4.w==w_4[j])].v, df4[(df4.u==u_4[i])& (df4.w==w_4[j])].B_v, 
                           '.-', lw=0.8, label='u=%.2f,w=%.2f'%(u_4[i], w_4[j]))
            ax0[0,2].plot(df4[(df4.u==u_4[i])& (df4.w==w_4[j])].v, df4[(df4.u==u_4[i])& (df4.w==w_4[j])].B_w, 
                           '.-', lw=0.8, label='u=%.2f,w=%.2f'%(u_4[i], w_4[j]))

for i in range(2,4):
    for j in range(len(w_4)):
            ax0[1,0].plot(df4[(df4.u==u_4[i])& (df4.w==w_4[j])].v, df4[(df4.u==u_4[i])& (df4.w==w_4[j])].B_u, 
                          '.-', lw=0.8, label='u=%.2f,w=%.2f'%(u_4[i], w_4[j]))
            ax0[1,1].plot(df4[(df4.u==u_4[i])& (df4.w==w_4[j])].v, df4[(df4.u==u_4[i])& (df4.w==w_4[j])].B_v, 
                           '.-', lw=0.8, label='u=%.2f,w=%.2f'%(u_4[i], w_4[j]))
            ax0[1,2].plot(df4[(df4.u==u_4[i])& (df4.w==w_4[j])].v, df4[(df4.u==u_4[i])& (df4.w==w_4[j])].B_w, 
                           '.-', lw=0.8, label='u=%.2f,w=%.2f'%(u_4[i], w_4[j]))


for i in [0,1]:
    for j in [0,1,2]:
        ax0[i,j].legend(fontsize=3.5, loc='upper right')
        ax0[i,0].set_ylabel('$\mathsf{B_u}\,(G)$')
        ax0[i,1].set_ylabel('$\mathsf{B_v}\,(G)$')
        ax0[i,2].set_ylabel('$\mathsf{B_w}\,(G)$')
        ax0[i,j].set_xlabel('$\mathsf{v}\,(cm)$')
fig0.tight_layout(pad=1, h_pad=1,rect=[0,0,0.9,0.9])
fig0.savefig('plots_gradients/gradient_v')
```
