# this notebook: test of the interpolation library


```python
from scipy.interpolate import Rbf
import numpy as np
import pandas as pd
%matplotlib notebook
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

```

```python
N = 10
xi = np.arange(N)
yi = 0.2 * np.arange(N)

df = pd.DataFrame([])
df['x'] = xi
df['y'] = yi
df['z'] = np.cos(np.sqrt(xi**2+yi**2))
```

```python
xl = np.linspace(1,N)
yl = 0.2*np.linspace(1,N)
xl_g, yl_g = np.meshgrid(xl, yl)
```

```python
rbf_z = Rbf(xi, yi, df.z[:10])
```

```python
zl = rbf_z(xl_g, yl_g)
```

```python
fig1 = plt.figure(facecolor='white', figsize=(5,5))

ax1 = fig1.add_subplot(111, projection='3d')
# ax2 = fig1.add_subplot(132, projection='3d')
# ax3 = fig1.add_subplot(133, projection='3d')
ax1.scatter3D(xi, yi, df.z,'o-')
ax1.scatter3D(xl_g, yl_g, zl)
```
